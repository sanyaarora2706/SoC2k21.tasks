```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
%matplotlib inline
```

```python
data=pd.read_csv("master.csv")
```

Discrete variables : country, year, sex, age, suicides_no, population, gdp, generation, gdp_per_capita

Continuous : suicides/100k, HDI


```python
data.columns
```

```python
#Discrete variables
```

```python
frequency=data["country"].value_counts(sort=False).to_dict()
x=[]
y=[]
for key in frequency:
    x.append(key)
    y.append(frequency[key])
fig = plt.figure(figsize=(16,4))
ax = fig.add_axes([0,0,1,1])
plt.title('Countries Frequency table')
ax.bar(x,y)
plt.setp(ax.get_xticklabels(), rotation=60, horizontalalignment='right', fontsize='x-small');
print('The mode is '+str(max(y)))
```

```python
frequency=data["year"].value_counts().to_dict()
x=[]
y=[]
for key in frequency:
    x.append(key)
    y.append(frequency[key])
fig = plt.figure(figsize=(16,4))
ax = fig.add_axes([0,0,1,1])
plt.title('Years Frequency table')
ax.bar(x,y);
print('The mode is '+str(max(y)))
```

```python
frequency=data["sex"].value_counts(sort=False).to_dict()
x=[]
y=[]
for key in frequency:
    x.append(key)
    y.append(frequency[key])
fig = plt.figure(figsize=(4,4))
ax = fig.add_axes([0,0,1,1])
plt.title('Sex Frequency table')
ax.bar(x,y);
print('The mode is '+str(max(y)))
```

```python
frequency=data["age"].value_counts().to_dict()
x=[]
y=[]
for key in frequency:
    x.append(key)
    y.append(frequency[key])
fig = plt.figure(figsize=(6,4))
ax = fig.add_axes([0,0,1,1])
plt.ylim(4600,4660)
plt.title('Age Frequency table')
ax.bar(x,y)
print('The mode is '+str(max(y)))
```

```python
frequency=data["suicides_no"].value_counts(sort=False).to_dict()
x=[]
y=[]
for key in frequency:
    x.append(key)
    y.append(frequency[key])
fig = plt.figure(figsize=(6,4))
ax = fig.add_axes([0,0,1,1])
plt.title('Suicides number frequency table')
plt.xlim(0,100)
ax.bar(x,y)
print('The mode is '+str(max(y)))
```

```python
frequency=data["population"].value_counts(sort=False).to_dict()
x=[]
y=[]
for key in frequency:
    x.append(key)
    y.append(frequency[key])
fig = plt.figure(figsize=(16,4))
ax = fig.add_axes([0,0,1,1])
plt.title('Population frequency table')
plt.xlim(0,0.8e6)
plt.ylim(0,5)
ax.bar(x,y)
print('The mode is '+ str(max(y)))
```
